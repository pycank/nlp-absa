```python
aspect_columns = [
    "AMBIENCE#GENERAL", "DRINKS#PRICES", "DRINKS#QUALITY",
    "DRINKS#STYLE&OPTIONS", "FOOD#PRICES", "FOOD#QUALITY",
    "FOOD#STYLE&OPTIONS", "LOCATION#GENERAL", "RESTAURANT#GENERAL",
    "RESTAURANT#MISCELLANEOUS", "RESTAURANT#PRICES", "SERVICE#GENERAL"
]
def add_binary_columns(example):
    for column in aspect_columns:
        new_column = f"is_{column}"
        example[new_column] = 1 if example[column] != 0 else 0
    return example

# Apply transformation to each split
preprocessed_datasets = preprocessed_datasets.map(add_binary_columns)

# Check the result
print(preprocessed_datasets["train"])
```

```python
def add_polarity_columns(example):
    for column in aspect_columns:
        polarity_key = f"polarity_{column}"
        value = example[column]
        if value == 1:
            example[polarity_key] = [1, 0, 0]
        elif value == 2:
            example[polarity_key] = [0, 1, 0]
        elif value == 3:
            example[polarity_key] = [0, 0, 1]
        else:
            example[polarity_key] = [0, 0, 0]  # For invalid or zero values
    return example

preprocessed_datasets = preprocessed_datasets.map(add_polarity_columns)
```

```python
from transformers import AutoModel
from torch import nn
import torch

class PBLAspect(nn.Module):
    def __init__(self, pretrained_model, num_labels=12):
        super(PBLAspect, self).__init__()
        self.embedder = AutoModel.from_pretrained(pretrained_model)
        self.num_labels = num_labels
        self.dropout1 = nn.Dropout1d(0.2)
        self.lstm = nn.LSTM(input_size=768, hidden_size=200, num_layers=1, bidirectional=True, batch_first=True)
        self.conv = nn.Conv1d(in_channels=400, out_channels=128, kernel_size=2, padding="valid")
        self.avg_pool = nn.AdaptiveAvgPool1d(1)
        self.max_pool = nn.AdaptiveMaxPool1d(1)
        self.dense1 = nn.Linear(256, 128)
        self.dropout2 = nn.Dropout(0.2)
        self.dense2 = nn.Linear(128, 64)
        self.dense3 = nn.Linear(64, num_labels)


    def forward(self, x, attention_mask):
        x = self.embedder(x, attention_mask=attention_mask)
        x = self.dropout1(x["last_hidden_state"])
        seq, (_, _) = self.lstm(x)
        x = seq.permute(0, 2, 1)
        x = self.conv(x)
        x = nn.ReLU()(x)
        gap = self.avg_pool(x).squeeze(2)
        gmp = self.max_pool(x).squeeze(2)
        x = torch.cat((gap, gmp), dim=1)
        x = self.dense1(x)
        x = nn.ReLU()(x)
        x = self.dropout2(x)
        x = self.dense2(x)
        x = nn.ReLU()(x)
        x = self.dense3(x)
        return x
```

```python
class FocalLoss(nn.Module):
    def __init__(self, alpha=0.25, gamma=2, reduction='mean'):
        super(FocalLoss, self).__init__()
        self.alpha = alpha
        self.gamma = gamma
        self.reduction = reduction
        self.bce_loss = nn.BCEWithLogitsLoss(reduction='none')

    def forward(self, inputs, targets):
        bce_loss = self.bce_loss(inputs, targets)
        pt = torch.exp(-bce_loss)
        focal_loss = self.alpha * (1 - pt) ** self.gamma * bce_loss
        if self.reduction == 'mean':
            return torch.mean(focal_loss)
```

```python
def calculate_accuracy(output, labels):
    correct = 0
    total = 0
    for i in range(output.shape[0]):
        for j in range(output.shape[1]):
            if output[i][j] >= 0.5 and labels[i][j] == 1:
                correct += 1
            elif output[i][j] < 0.5 and labels[i][j] == 0:
                correct += 1
            total += 1
    return correct, total
```

```python
from torch.utils.data import DataLoader
from torch.optim import AdamW
from torch.optim.lr_scheduler import CosineAnnealingLR
from tqdm import tqdm
import torch
```

```python


def train(model, optimizer, scheduler, train_dataloader, val_dataloader, device):
    loss_fn = FocalLoss(gamma=2)

    model.to(device)  # Move the model to GPU
    model.train()
    total_correct = 0
    total_samples = 0

    for batch in tqdm(train_dataloader, total=len(train_dataloader), desc='Training'):
        optimizer.zero_grad()
        input_ids = batch['input_ids'].to(device)  # Move inputs to GPU
        attention_mask = batch['attention_mask'].to(device)
        labels = torch.stack([batch[col] for col in label_columns], dim=1).to(device)

        # Forward pass
        output = model(input_ids, attention_mask=attention_mask)
        # print(output)
        # print(labels)
        batch_loss = loss_fn(output, labels.float())

        # Backward pass
        batch_loss.backward()
        optimizer.step()
        scheduler.step()
        correct, samples = calculate_accuracy(nn.Sigmoid()(output), labels)
        total_correct += correct
        total_samples += samples
    accuracy = total_correct / total_samples * 100
    print(f"Training Loss: {batch_loss.item()}, Accuracy: {accuracy:.2f}%")


    # Validation
    model.eval()
    total_correct = 0
    total_samples = 0
    total_val_loss = 0.0

    with torch.no_grad():
        for batch in tqdm(val_dataloader, total=len(val_dataloader), desc='Validation'):
            input_ids = batch['input_ids'].to(device)  # Move inputs to GPU
            attention_mask = batch['attention_mask'].to(device)
            labels = torch.stack([batch[col] for col in label_columns], dim=1).to(device)  # Move labels to GPU

            # Forward pass
            output = model(input_ids, attention_mask=attention_mask)
            val_loss = loss_fn(output, labels.float())
            total_val_loss += val_loss.item()

            # Accuracy calculation
            correct, samples = calculate_accuracy(nn.Sigmoid()(output), labels)
            total_correct += correct
            total_samples += samples

    avg_val_loss = total_val_loss / len(val_dataloader)
    accuracy = total_correct / total_samples * 100
    print(f"Validation Loss: {avg_val_loss:.4f}, Accuracy: {accuracy:.2f}%")
```

```python

epoch = 5
batch_size = 8
model = PBLAspect(PRETRAINED_MODEL)
model.to('cuda')
train_dataloader = DataLoader(preprocessed_datasets['train'], batch_size=batch_size, shuffle=True)
val_dataloader = DataLoader(preprocessed_datasets['val'], batch_size=batch_size, shuffle=False)
optimizer = AdamW(model.parameters(), lr=1e-4)
scheduler = CosineAnnealingLR(optimizer, T_max=epoch)
for _ in range(epoch):
    train(model, optimizer, scheduler, train_dataloader, val_dataloader, 'cuda')
```

```python
class PBLPolarity(nn.Module):
    def __init__(self, pretrained_model, num_labels=3):
        super(PBLPolarity, self).__init__()
        self.embedder = AutoModel.from_pretrained(pretrained_model)
        self.num_labels = num_labels
        self.dropout1 = nn.Dropout1d(0.2)
        self.lstm = nn.LSTM(input_size=768, hidden_size=200, num_layers=1, bidirectional=True, batch_first=True)
        self.conv = nn.Conv1d(in_channels=400, out_channels=128, kernel_size=2, padding="valid")
        self.avg_pool = nn.AdaptiveAvgPool1d(1)
        self.max_pool = nn.AdaptiveMaxPool1d(1)
        self.dense1 = nn.Linear(256, 128)
        self.dropout2 = nn.Dropout(0.2)
        self.dense2 = nn.Linear(128, 64)
        self.dense3 = nn.Linear(64, num_labels)


    def forward(self, x, attention_mask):
        x = self.embedder(x, attention_mask=attention_mask)
        x = self.dropout1(x["last_hidden_state"])
        seq, (_, _) = self.lstm(x)
        x = seq.permute(0, 2, 1)
        x = self.conv(x)
        x = nn.ReLU()(x)
        gap = self.avg_pool(x).squeeze(2)
        gmp = self.max_pool(x).squeeze(2)
        x = torch.cat((gap, gmp), dim=1)
        x = self.dense1(x)
        x = nn.ReLU()(x)
        x = self.dropout2(x)
        x = self.dense2(x)
        x = nn.ReLU()(x)
        x = self.dense3(x)
        # x = nn.Softmax(dim=1)(x)
        return x
```

```python
def train_polarity(aspect, model, optimizer, scheduler, train_dataloader, val_dataloader, device):
    loss_fn = nn.CrossEntropyLoss()

    model.to(device)  # Move the model to GPU
    model.train()
    total_correct = 0
    total_samples = 0
    for batch in tqdm(train_dataloader, total=len(train_dataloader), desc='Training'):
        optimizer.zero_grad()
        input_ids = batch['input_ids'].to(device)  # Move inputs to GPU
        attention_mask = batch['attention_mask'].to(device)
        labels = batch[f'polarity_{aspect}'].to(device)  # Move labels to GPU

        # Forward pass
        output = model(input_ids, attention_mask=attention_mask)
        batch_loss = loss_fn(output, labels.float())
        correct = (torch.argmax(output, dim=1) == labels.argmax(dim=1)).sum().item()
        total_correct += correct
        total_samples += len(labels)
        # Backward pass
        batch_loss.backward()
        optimizer.step()
        scheduler.step()
    print(batch_loss.item())
    accuracy = total_correct / total_samples * 100
    print(f'Training Loss: {batch_loss.item()}, Accuracy: {accuracy:.2f}%')

    # Validation
    model.eval()
    total_correct = 0
    total_samples = 0
    total_val_loss = 0.0
    with torch.no_grad():
        for batch in tqdm(val_dataloader, total=len(val_dataloader), desc='Validation'):
            input_ids = batch['input_ids'].to(device)  # Move inputs to GPU
            attention_mask = batch['attention_mask'].to(device)
            labels = batch[f'polarity_{aspect}'].to(device)

            output = model(input_ids, attention_mask=attention_mask)
            val_loss = loss_fn(output, labels.float())
            total_val_loss += val_loss.item()

            correct = (torch.argmax(output, dim=1) == labels.argmax(dim=1)).sum().item()
            total_correct += correct
            total_samples += len(labels)

    avg_val_loss = total_val_loss / len(val_dataloader)
    accuracy = total_correct / total_samples * 100
    print(f'Validation Loss: {avg_val_loss:.4f}, Accuracy: {accuracy:.2f}%')

```

```python
epoch = 3
batch_size = 8
aspects = aspect_columns
for aspect in aspects:
    model = PBLPolarity(PRETRAINED_MODEL)
    # Use Dataset.filter to filter the dataset based on the 'is_{aspect}' column
    train_dataloader = DataLoader(preprocessed_datasets['train'].filter(lambda example: example[f'is_{aspect}'] != 0), batch_size=batch_size, shuffle=True)
    val_dataloader = DataLoader(preprocessed_datasets['val'].filter(lambda example: example[f'is_{aspect}'] != 0), batch_size=batch_size, shuffle=False)
    optimizer = AdamW(model.parameters(), lr=1e-4)
    scheduler = CosineAnnealingLR(optimizer, T_max=epoch)
    for _ in range(epoch):
        train_polarity(aspect, model, optimizer, scheduler, train_dataloader, val_dataloader, 'cuda')
    torch.save(model.state_dict(), f'/content/drive/MyDrive/WebMining/pblp-restaurant-{aspect}.pt')
    del model
    del optimizer
    del scheduler
```

```python
# evaluation
test_dataloader = DataLoader(preprocessed_datasets['test'], batch_size=8, shuffle=False)

stage1 = PBLAspect(PRETRAINED_MODEL)
stage1.load_state_dict(torch.load('/content/drive/MyDrive/WebMining/pbla-restaurant.pt'))
stage1.to('cuda')
stage1.eval()
stage2 = []
for aspect in aspect_columns:
    stage2.append(PBLPolarity(PRETRAINED_MODEL))
    stage2[-1].load_state_dict(torch.load(f'/content/drive/MyDrive/WebMining/pblp-restaurant-{aspect}.pt'))
    stage2[-1].to('cuda')
    stage2[-1].eval()
```

```python
total_tp = {}
total_tp["aspect"], total_tp["polarity"] = 0, 0
total_fp = {}
total_fp["aspect"], total_fp["polarity"] = 0, 0
total_fn = {}
total_fn["aspect"], total_fn["polarity"] = 0, 0
```

```python

with torch.no_grad():
    for batch in tqdm(test_dataloader, total=len(test_dataloader), desc='Evaluating'):
        input_ids = batch['input_ids'].to('cuda')
        attention_mask = batch['attention_mask'].to('cuda')
        output = nn.Sigmoid()(stage1(input_ids, attention_mask=attention_mask))
        labels = torch.stack([batch[col] for col in label_columns], dim=1).to('cuda') # original labels (batch_size, num_aspects)

        true_positive = ((output > 0.5) & (labels == 1)).sum(dim=0)
        false_positive = ((output > 0.5) & (labels == 0)).sum(dim=0)
        false_negative = ((output <= 0.5) & (labels == 1)).sum(dim=0)
        total_tp["aspect"] += true_positive.sum().item()
        total_fp["aspect"] += false_positive.sum().item()
        total_fn["aspect"] += false_negative.sum().item()

        # Predictions for each aspect
        aspect_predictions = []
        for j in range(output.shape[1]):  # Iterate through aspects
            aspect_output = output[:, j]  # Get output for the current aspect
            aspect_indices = aspect_output >= 0.5  # Indices where aspect is predicted

            # Get polarity predictions for the selected indices using stage2 models
            selected_input_ids = input_ids[aspect_indices]
            selected_attention_mask = attention_mask[aspect_indices]
            if len(selected_input_ids) > 0:
                polarity_output = stage2[j](selected_input_ids, attention_mask=selected_attention_mask)
                polarity_predictions = polarity_output.argmax(dim=1) + 1  # +1 to map 0, 1, 2 to 1, 2, 3
            else:
                polarity_predictions = torch.tensor([], device='cuda', dtype=torch.int64)

            # Create a tensor to store the predictions for the current aspect
            current_aspect_predictions = torch.zeros(output.shape[0], dtype=torch.int64, device='cuda')
            current_aspect_predictions[aspect_indices] = polarity_predictions
            aspect_predictions.append(current_aspect_predictions)

        # Combine the predictions for all aspects
        output = torch.stack(aspect_predictions, dim=1)  # (batch_size, num_aspects)

        labels = torch.stack([batch[col] for col in aspect_columns], dim=1).to('cuda') # original labels (batch_size, num_aspects)

        true_positive = ((output > 0) & (labels == output)).sum().item()
        false_positive = ((output > 0) & (labels == 0)).sum().item()
        false_negative = ((output == 0) & (labels > 0)).sum().item()
        total_tp["polarity"] += true_positive
        total_fp["polarity"] += false_positive
        total_fn["polarity"] += false_negative

    precision_aspect = total_tp["aspect"] / (total_tp["aspect"] + total_fp["aspect"])
    recall_aspect = total_tp["aspect"] / (total_tp["aspect"] + total_fn["aspect"])
    f1_aspect = 2 * precision_aspect * recall_aspect / (precision_aspect + recall_aspect)

    precision_polarity = total_tp["polarity"] / (total_tp["polarity"] + total_fp["polarity"])
    recall_polarity = total_tp["polarity"] / (total_tp["polarity"] + total_fn["polarity"])
    f1_polarity = 2 * precision_polarity * recall_polarity / (precision_polarity + recall_polarity)
print("\n")
print(f"Aspect Precision: {precision_aspect}, Aspect Recall: {recall_aspect}, Aspect F1: {f1_aspect}")
print(f"Polarity Precision: {precision_polarity}, Polarity Recall: {recall_polarity}, Polarity F1: {f1_polarity}")
```
